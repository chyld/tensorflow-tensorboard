```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
X1 = np.array([
    [1, 1250],
    [1, 1700],
    [1, 1400],
    [1, 1900]
])

y1 = np.array([111000, 222000, 155000, 277000])
```

```python
def predict(X, weights):
#     print('X:', X, 'weights:', weights)
    return X.dot(weights)

def cost(m, b):
#     print('m:', m, 'b:', b)
    y_hat = predict(X1, np.array([b, m]))
    return np.sum((y1 - y_hat) ** 2)
```

```python
m = np.arange(   -300,    300,     5)
b = np.arange(-220000, 220000,  5000)

fig = plt.figure(figsize=(20,7))

ax = fig.add_subplot(121, projection='3d')
ax.view_init(elev=30.0, azim=90)

X, Y = np.meshgrid(m, b)

zs = np.array([cost(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)
ax.plot_surface(X, Y, Z,cmap='hot')
```
